```python
def lat_long_finder():

    cities = ["Delhi", "Gurugram", "Mumbai", "Chennai", "Bengaluru", "Hyderabad", 
              "Kolkata", "Pune", "Jaipur", "Ahmedabad", "Lucknow", "Surat", 
              "Chandigarh", "Bhopal", "Indore", "Patna", "Coimbatore", 
              "Kochi", "Nagpur", "Visakhapatnam"]
    
    result = []

    url = 'https://api.openweathermap.org/geo/1.0/direct'

    for city in cities:
        params = {
            'q' : city,
            'limit': 1,
            'appid' : api_key
        }
        coordinates = requests.get(url,params)

        data = coordinates.json()

        result.append(data[0])
        

    df = pd.DataFrame(result)

    return df.to_csv("latitude_longitude_date.csv", index = False)
```

```python
coordinate_data = pd.read_csv("latitude_longitude_date.csv")

coordinate_data
```

```python
def city_data():

    coord_data = []

    for index,row in coordinate_data.iterrows():
        lat = row['lat']
        lon = row['lon']

        coord_data.append((lat,lon))

    return coord_data
    
```

```python
city_data()
```

```python
def weather_data():

    city_coord = city_data()
    
    weather_url = 'https://api.openweathermap.org/data/2.5/weather'

    result_weather = []

    for i in city_coord:
        x = 0
        y = 1
        params = {
            'lat': i[x],
            'lon': i[y],
            'appid': api_key
        }

        weather = requests.get(weather_url,params)

        weather_data = weather.json()

        result_weather.append(weather_data)

        x +=1
        y +=1

    # df_weather = pd.DataFrame(result_weather)

    df_flatten = pd.json_normalize(result_weather)

    df_weather = pd.json_normalize(df_flatten['weather'].apply(lambda x:x[0]))

    df_merge = pd.concat([df_weather,df_flatten.drop('weather', axis = 1)], axis = 1)

    saved_csv = df_merge.to_csv("weather_data.csv", index = False)

    return saved_csv
        
```

```python
def worst_cities():
    columns_extract = ['name','visibility']
    
    df_visbility = y[columns_extract]
    
    df_visbility
    
    quartile = df_visbility['visibility'].quantile([0.25,0.5,0.75,0.9])
    
    bins = [0, quartile[0.25],quartile[0.5],quartile[0.75],quartile[0.9]]
    labels = ['Bottom quartile', 'Lower-middle', 'Upper-middle', 'Top quartile']
    
    df_visbility['quartile'] = pd.cut(df_visbility['visibility'],bins = bins, labels = labels, include_lowest = True)

    df_sorted = df_visbility.sort_values(by = 'visibility')

    bottom_df = df_sorted.loc[df_sorted['quartile'] == 'Bottom quartile']

    worst_city_list = bottom_df['name'].tolist()

    print("The Worst Cities to live in are: ")
    print(worst_city_list)
```
