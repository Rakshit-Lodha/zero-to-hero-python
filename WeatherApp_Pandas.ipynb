```python
def lat_long_finder():

    cities = ["Delhi", "Gurugram", "Mumbai", "Chennai", "Bengaluru", "Hyderabad", 
              "Kolkata", "Pune", "Jaipur", "Ahmedabad", "Lucknow", "Surat", 
              "Chandigarh", "Bhopal", "Indore", "Patna", "Coimbatore", 
              "Kochi", "Nagpur", "Visakhapatnam"]
    
    result = []

    url = 'https://api.openweathermap.org/geo/1.0/direct'

    for city in cities:
        params = {
            'q' : city,
            'limit': 1,
            'appid' : api_key
        }
        coordinates = requests.get(url,params)

        data = coordinates.json()

        result.append(data[0])
        

    df = pd.DataFrame(result)

    return df.to_csv("latitude_longitude_date.csv", index = False)
```

```python
def city_data():

    coord_data = []

    for index,row in coordinate_data.iterrows():
        lat = row['lat']
        lon = row['lon']

        coord_data.append((lat,lon))

    return coord_data
    
```
