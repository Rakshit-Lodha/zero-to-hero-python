```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/Rakshit.Lodha/Desktop/McDonalds_Financial_Statements.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
df[df.columns[1:len(df)-1]].describe()
```

```python
df.sort_values(by=['Year'], inplace = True)

df
```

```python
df['Debt to Equity'] = df['Total liabilities ($B)']/df['Shares Outstanding ($B)']
df['% Change in Revenue'] = (df['Revenue ($B)'].pct_change())*100
df['% Change in Market Cap'] = (df['Market cap ($B)'].pct_change())*100

df['Year'] = df.index

df
```

```python
df.plot(x ='Year', y= ['% Change in Revenue','% Change in Market Cap'])

plt.show
```

```python
max_date = df['% Change in Revenue'].nlargest(1)
min_date = df['% Change in Revenue'].nsmallest(1)

print(f"Sharing the details of largest and smallest increase in revenue:\nMax Revenue:{max_date}\nMin Revenue: {min_date}")
```

```python
df_before2008 = df[df['Year'] < '2008']
df_after2008 = df[df['Year'] >= '2008']

print(f"Revenue Before 2008: {df_before2008['Revenue ($B)'].mean()}\nRevenue After 2008: {df_after2008['Revenue ($B)'].mean()}")
```
