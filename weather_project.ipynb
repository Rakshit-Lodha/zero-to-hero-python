```python
import requests
```

```python
url = 'http://api.openweathermap.org/geo/1.0/direct'

params = {
    'q': 'Gurugram', 
    'limit': 1,
    'appid': '7956fb1c004aa5f5932d28709f0d75aa'
}


coordinate = requests.get(url,params)

loop = coordinate.json()
```

```python
for i in loop:
    lat = i['lat']
    lon = i['lon']

    print(f"Lat: {lat}, Lon: {lon}")
```
