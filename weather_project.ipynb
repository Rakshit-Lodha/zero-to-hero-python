```python
import requests
```

```python
def lat_long_finder():
    
    city = input("Which city would you like to search details of? ")
    
    lat_long_url = 'http://api.openweathermap.org/geo/1.0/direct'
    
    params = {
        'q': city, 
        'limit': 1,
        'appid': api_key
    }
    
    
    coordinate = requests.get(lat_long_url,params)

    loop = coordinate.json()

    return loop
```

```python
def lat_long_checker():
    while True:
        details = lat_long_finder()
        if details == []:
            print("City Doesn't Exist")
        else:
            return details
            break
```

```python
def city_data():

    city_coordinates = lat_long_checker()

    for i in city_coordinates:
        lat = i['lat']
        lon = i['lon']
    
    weather_url = 'https://api.openweathermap.org/data/2.5/weather'
    params = {
    'lat': lat,
    'lon': lon,
    'appid': api_key
    }
    
    weather_details = requests.get(weather_url,params)
    
    second_call = weather_details.json()

    current_temperature = second_call['main'].get('temp')
    min_temp = second_call['main'].get('temp_min')
    max_temp = second_call['main'].get('temp_max')
    city = second_call['name']
    country = second_call['sys'].get('country')
    
    return print(f'Here are the details of {city},{country}:\nCurrent Temperature: {current_temperature}\nMin Temperature: {min_temp}\nMax Temperature: {max_temp}')
```

```python
city_data()
```
